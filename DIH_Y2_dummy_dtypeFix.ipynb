```python
df5=df4
df5.columns
```

```python
df5 = df5.drop(['AgeAtFirstClaim' ,'ProviderID','Vendor','PCP'], axis=1)
#yourdf.drop(['columnheading1', 'columnheading2'], axis=1, inplace=True)
```

```python
df5.columns
df5.shape
```

```python
#trying an approach of taking mean of all
pivot1 = pd.pivot_table(df4,index=["MemberID"])
pivot1.head()
```

```python
pivot1.shape
```

```python
#doing sum of all dummy variables and ignoring 4 int variables where we need to do a mean
pivot0= pd.pivot_table(df5,index=['MemberID'],
                     aggfunc=[np.sum])
pivot0.head()
```

```python
#mean of Age 
pivot0_a= pd.pivot_table(df4,index=['MemberID'],values=['AgeAtFirstClaim'],
                     aggfunc=[np.mean])
pivot0_a.head()
```

```python
result = pd.concat([pivot0, pivot0_a], axis=1)
result.head()
```

```python
#distinct count of 3 integer variables  
pivot0_b= pd.pivot_table(df4,index=['MemberID'],values=['ProviderID','Vendor','PCP'],
                     aggfunc=lambda x: len((x.dropna()).unique()))

pivot0_b.head(20)
```

```python
###### this is the final aggregated df to export to r

result1 = pd.concat([result, pivot0_b], axis=1)
result1.head(20)
```

```python
result1.to_csv('out_Agg_dummy_Y2_r01.csv',index=True)
```
